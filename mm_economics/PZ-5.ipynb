Практическое занятие №4\
**Модель экономического роста Слоу**


```python
# Производственная функция модели Слоу

# Y = A*K&alpha*(L*E)^(1-alpha)

# 1-й момент модели Солоу
# L - труд (численность экономически активного населения)
# E - эффективность труда (отражает влияние на фактор труд НТП)
# E - ненаблюдаемая перменная
```

```python
# 2-й момент модели Солоу:
# ПФ в модели Солоу - линейно-однородная: F(m*K, m*L) = m*F(K, L)
```

```python
# 3-й момент:
# Y/(LE) = F(K, LE) = F(K/(LE), 1) m = 1/(LE) => y = f(k), k = K/(LE)
```

```python
# 4-й момент:
# Исходная функция: Y = A*K^alpha(LE)^(1-alpha) \ LE
# Функция Кобба-Дугласа в модели Солоу принимает вид:
# y = A*k^alpha,
# A - параметр общей факторной производительности (отражает влияние на
# валовой продукт страны дополнительных факторов (кроме труда и капитала),
# например, НТП)
# alpha - эластичность выпуска по капиталу (показывает, на сколько процентов
# увеличится среднедушевой выпуск (доход) при увеличении капитала на 1%)
```

**Модель экономического роста - 1**

```python
import sympy as sp
```

```python
A, alpha, K, L, E = sp.symbols('A alpha K L E')
```

```python
Y = A*K**alpha*(L*E)**(1-alpha)
```

```python
sp.simplify(Y/(L*E))
```
