```python
# Условие равновесия на фин рынке
# S + CF(r) = I(r),     I'(r) < 0,  CF'(r) > 0

# Условие равновесия на товарном рынке
# S = Y - C - G - NX(eps) = I(r) => Y = C + I(r) + G + NX(eps)
```

```python
# S - I(r) = NX(eps), S < I(r) И NX(eps) < 0

# S - I(r) = Ex(eps) - Im(eps)
# Im(eps) = Ex(eps) - (S - I(r)) = Ex(eps) + CF(r)

# Im(eps) = Ex(eps) + CF(r)
```

```python
import sympy as sp
```

```python
# Эндогенные переменные
C, I, NX, CF, r, eps = sp.symbols('c, I, NX, CF, r, eps')
```

```python
# Экзогенные переменные
Y, T, G = sp.symbols('T, G, Y')
```

```python
# Параметры модели
a0, a1, b0, b1, c0, c1, d0, d1 = sp.symbols('a0, a1, b0, b1, c0, c1, d0, d1')
```

```python
sp.init_printing()
A = sp.solve([C - a0 - a1*(Y - T),
             I - b0 - b1*r,
             NX - c0 - c1*eps,
             CF - d0 - d1*r,
             Y - C - I - G - NX,
             NX - CF], [C, I, NX, CF, eps, r], dict = True)[0]; A
```

```python
# Сдерживающая политикаЖ: Delta_T > 0, Delta_G < 0
```

```python
Delta_T, Delta_G = sp.symbols('Delta_T, Delta_G')
```

```python
Delta_C = A[C].diff(T)*Delta_T; Delta_C
```

```python
# Delta_C < 0, т.е. повышение налогов приведет долгосрочном периоде
# к снижению потребительских расходов
```

```python
Delta_CF = A[CF].diff(T)*Delta_T + A[CF].diff(G)*Delta_G; Delta_CF
```

```python
# (-Delta_G*d1 + Delta_T*a1*d1)/(b1 + d1)
# -Delta_G*d1 + Delta_T*a1*d1 > 0
# Если b1 + d1 > 0 => Delta_CF > 0
# Если b1 + d1 < 0 => Delta_CF < 0
```

```python
Delta_NX = A[NX].diff(T)*Delta_T + A[NX].diff(G)*Delta_G; Delta_NX
```

```python
# Реакция NX оценивается анaлогично реакции CF
```

```python
Delta_eps = A[eps].diff(T)*Delta_T + A[eps].diff(G)*Delta_G; Delta_eps
```

```python
Delta_r = A[r].diff(T)*Delta_T + A[r].diff(G)*Delta_G; Delta_r
```

```python
Delta_I = A[I].diff(T)*Delta_T + A[I].diff(G)*Delta_G; Delta_I
```
