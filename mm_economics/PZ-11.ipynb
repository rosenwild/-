```python
# p_in увеличивается при e=const и p_out = const => eps увеличивается
# Вопрос: как это отразится на величине чистого экспорта NX?
# NX = Ex - Im => чистый экспорт будет снижаться, так как
# с повышением цены экспортируемого товара спрос внешнего мира
# на этот товар будет снижаться, следовательно, Ex будет снижаться
# (на импорт не повлияет, т.к. повышается цена на наш товар) =>
# NX тоже будет снижаться
```

```python
# Равновесие на валютном рынке:
# S - I(r*) + NX(eps)
# S - I(r*) = const, S = Y - C - G = const, r* - мировая ставка процента
# В МОЭ макроэкономическое равновесие устанавливается благодаря гибкости eps
```

```python
# S - I(r*) - EX = - Im
# Ex  - (S - I(r*)) = Im,    S - I(r*) > 0
```

```python
# Из равновесия валютного рынка следует равновесие товарного рынка:
# S - I(r*) = NX(eps) => (Y - C - G) - I(r*) = NX(eps) =>
# Y = C  + G + I(r*) + NX(eps) - условие равновесия на товарном рынке
```

# 3.

image.png

```python
C, I, NX, eps = sp.symbols('C I NX epsilon') # эндогенные (целевые) переменные
T, G, Y, r = sp.symbols('T G Y r') # экзогенные переменные (рассчитываются за пределами модели)
a0, a1, b0, b1, c0, c1 = sp.symbols('a0 a1 b0 b1 c0 c1') # параметры
# b1 < 0, c1 < 0, 0 < a0 < 1, a1 > 0, b0 > 0
```

```python
sp.init_printing()
A = sp.solve([Y - C - I - G - NX, C - a0 - a1*(Y - T),
              I - b0 - b1*r, NX - c0 - c1*eps], [C, I, NX, eps], dict = True)[0]; A
```

# 4.

image.png

```python
# на фоне снижения мировой ставки процента
```

```python
Delta_T, Delta_G, Delta_r = sp.symbols('Delta_T, Delta_G, Delta_r')
# Delta_T < 0, Delta_G > 0, Delta_r < 0
```

```python
# Delta_eps - ?
sp.expand(A[eps])
```

```python
# реакция реального валютного курса:
Delta_eps = A[eps].diff(T)*Delta_T + A[eps].diff(G)*Delta_G + A[eps].diff(r)*Delta_r; Delta_eps
```

```python
# Delta_eps > 0
```

```python
# Delta_NX - ?
sp.expand(A[NX])
```

```python
Delta_NX = A[NX].diff(T)*Delta_T + A[NX].diff(G)*Delta_G + A[NX].diff(r)*Delta_r; Delta_NX
```

```python
# Delta_NX < 0
```

```python
# Delta_C - ?
Delta_C = A[C].diff(T)*Delta_T; Delta_C
```

```python
# Delta_C > 0
```

```python
# Delta_I - ?
Delta_I = A[I].diff(r)*Delta_r; Delta_I
```

```python
# Delta_I > 0
```

```python
# Delta_S - ?   S = Y - C - G
```

```python
S = Y - A[C] - G; S
```

```python
Delta_S = S.diff(T)*Delta_T; Delta_S
```

```python
# Delta_S < 0
```

```python

Spr = (Y - T) - A[C]; Spr
```

```python
Delta_Spr = Spr.diff(T)*Delta_T; Delta_Spr
```

```python
# Delta_Spr > 0
```
