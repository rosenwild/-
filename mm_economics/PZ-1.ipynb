```python
# Т.к. экономика закрытая, условие равновесия на товарном рынке имеет вид:
# Y = C + I + G - первое уравнение
```

```python
# T = 0.3*Y - второе уравнение
```

```python
# Tr = 1/3*T - третье уравнение
```

```python
# D = G - T + Tr - четвертое уравнение
```

```python
import sympy as sp
```

```python
C = 2200; I = 600; D = 80
```

```python
Y, G, T, Tr = sp.symbols(' Y G T Tr')
```

```python
Sol1 = sp.solve([Y - C - I - G, T - 0.3*Y, Tr - 1/3*T, D - G + T - Tr], [Y, G, T, Tr], dict = True)[0]; Sol1
```

```python
Tn = Sol1[T] - Sol1[Tr]; Tn
```

```python
D = Sol1[G] - Tn; D
```

```python
# Y = C + I + G + Ex - C_f - I_f - G_f = C + I + G + Ex - Im = C + I + G + NX
```

```python
# Y = C + I + G + NX - первое уравнение
```

```python
# NX = Ex - Im - второе уравнение
```

```python
# L = T - Tr + Sp + Im - изъятия - часть дохода, которая не стимулирует рост
# национальной экономики - третье уравнение
```

```python
# Sp = Y - T + Tr - C - четвертое уравнение
```

```python
# Sl = T - Tr - G - сальдо государственного бюджета - пятое уравнение
```

```python
C, G, Im, Sp, Sl = sp.symbols('C G Im Sp Sl')
```

```python
Sol3 = sp.solve([Y - C - I - G - NX, NX - Ex + Im, L - T + Tr - Sp - Im, Sp - Y + T - Tr + C, Sl - T + Tr + G], [C, G, Im, Sp, Sl], dict = True)[0]; Sol3
```
